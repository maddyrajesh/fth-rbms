```python
from itertools import combinations
from operator import itemgetter
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
customer_df = pd.read_csv('customerdata.csv',low_memory=False)
customer_df.head()
```

```python
item_purchase_df = customer_df.groupby(['CUSTOMER_ID','PRODUCT_NAME']).size().reset_index(name='counts')
```

```python
item_purchase_df
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder = LabelEncoder()
item_purchase_df['PRODUCT_NAME'] = labelEncoder.fit_transform(item_purchase_df['PRODUCT_NAME'])
```

```python
item_purchase_df.head()
```

```python
item_purchase_df['counts'] = (item_purchase_df['counts']-item_purchase_df['counts'].min())/(item_purchase_df['counts'].max()-item_purchase_df['counts'].min())
```
