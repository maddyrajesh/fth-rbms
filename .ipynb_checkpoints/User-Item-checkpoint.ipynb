```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
customer_df = pd.read_csv('customerdata.csv',low_memory=False)
customer_df.head()
```

```python
item_number=pd.DataFrame(customer_df.groupby('CUSTOMER_ID')["PRODUCT_NAME"].apply(list))
item_number
```

```python
def to_1D(series):
    return pd.Series([x for _list in series for x in _list])
```

```python
unique_items = to_1D(item_number["PRODUCT_NAME"]).value_counts()
```
